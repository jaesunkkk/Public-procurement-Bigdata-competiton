### 날짜 변수 변환&추가

- 납품기한일자에서 계약일자또는 최초계약일자를 빼서 납품하는데 걸리는 기간 변수를 만든다. 이때 train 데이터의 세가지 날짜 변수는 int형 데이터 타입이므로 이를 to_datetime함수를 이용하여 
날짜형으로 바꾼다. 이후 분석에서는 이 기간 변수를 사용할 때 날짜연산 대신 숫자연산이 필요하기 때문에 다시 int형으로 변환해주고, 존재하는 음수값은 삭제한다.

````python
train['계약(납품요구)일자']=pd.to_datetime(train['계약(납품요구)일자'],format='%Y%m%d')
train['최초계약(납품요구)일자']=pd.to_datetime(train['최초계약(납품요구)일자'],format='%Y%m%d')
train['납품기한일자']=pd.to_datetime(train['납품기한일자'],format='%Y%m%d')
````

````python
train['납품기한일자-계약(납품요구)일자']=train['납품기한일자']-train['계약(납품요구)일자']
train['납품기한일자-최초계약(납품요구)일자']=train['납품기한일자']-train['최초계약(납품요구)일자']
````

````python
#추가:datetime-> int형변환
train['납품기한일자-계약(납품요구)일자']=train['납품기한일자-계약(납품요구)일자'].dt.days
train['납품기한일자-최초계약(납품요구)일자']=train['납품기한일자-최초계약(납품요구)일자'].dt.days
````

### 수량(ton)변수 추가 


- train 데이터에는 같은 단위이지만 표현이 다른 값들과 여러 단위가 존재하므로 수월한 예측을 위해 수량을 ton으로 통일한다. train데이터의 단위들을 살펴보면 톤에 해당하는 단위에는 톤, 백, ton, 톤백, T, 톤백(톤), 톤(25kg백), 25kg, 40포, 1톤의 입력값이 있고 포에는 포, 25kg포, 포대(포), 10kg/포의 입력값, kg에는 kg, 키로, 그 외에는 L, 식, 개의 값이 있다. 톤에 해당하는 단위들은 변형없이 수량변수에 추가하고, kg단위는 0.001을 곱합니다. 개와 포의 경우 품목을 확인했을 때, 염화칼슘 제설제의 규격인 포나 마대의 개수를 의미하는 것을 알 수 있으므로 명시되어있지 않은것은 1000kg로 처리 그 외는 명시되어 있는대로 처리한다. L의 경우 부피단위이므로 밀도(2.15ton/m3)를 이용하여 ton 수량을 구해보았다. L와 톤을 이용한 평균단가를 비교하면 L의 평균단가는 185568, 톤의 평균단가는 218449이므로 L의 평균단가를 밀도를 이용해 톤에 맞춰 변형시켰을때 86.3109이 나오므로 금액을 이용하여 L를 톤으로 변경하는 방법이 더 좋은 방법이라고 판단하였다. 

- 1L당->ton: 185568/(0.001*2.15)=86.31069
- 1톤당 : 218449

- 금액을 이용한 L단위의 변환은 톤 단위 평균 단가로 총금액을 나누는 것으로 했다. 
- 정의가 불분명한 '식'의 단위도 이와같은 방법을 이용하여 변환했다.




(x) 톤백과 톤(25kg백) 같은것..? = 같다고 봤음 (단위는 1톤인데 25kg 백에 담겨있다고 해석)

### IQR(Inter Quantile Range) 방식으로 이상치 제거 

````python
import numpy as np

def get_outlier(df=None, column=None, weight=1.5):
  quantile_25 = np.percentile(df[column].values, 25)
  quantile_75 = np.percentile(df[column].values, 75)

  IQR = quantile_75 - quantile_25
  IQR_weight = IQR*weight
  
  lowest = quantile_25 - IQR_weight
  highest = quantile_75 + IQR_weight
  
  outlier_idx = df[column][ (df[column] < lowest) | (df[column] > highest) ].index
  return outlier_idx

# 함수 사용해서 이상치 값 삭제
outlier_idx = get_outlier(df=train, column='수량(ton)', weight=1.5)
train.drop(outlier_idx, axis=0, inplace=True)
````

describe() 함수를 이용해 수량(ton) 변수의 분포를 확인한 결과, 꼬리가 긴 분포를 가지는 것으로 확인되었다. 따라서 IQR(Inter Quantile Range) 방식으로 이상치를 판단하여 해당 index를 반환하는 get_outlier 함수를 만들었고, 이 함수를 train 데이터에 적용하여 이상치를 제거하였다.

````python
def draw_blockMap(blockedMap, targetData, title, color ):
    whitelabelmin = (max(blockedMap[targetData]) - min(blockedMap[targetData]))*0.25 + min(blockedMap[targetData])
    
    datalabel = targetData
    
    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])
    mapdata = blockedMap.pivot(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(8, 13))
    plt.title(title)
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=color, edgecolor='#aaaaaa',
linewidth=0.5)
    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
# 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시 (중구, 서구)
        if row['광역시도'].endswith('시') and not row['광역시도'].startswith('세종'):
            dispname = '{}\n{}'.format(row['광역시도'][:2], row['행정구역'][:-1]) 
            if len(row['행정구역'])<= 2:
              dispname += row['행정구역'][-1] 
        else:
            dispname = row['행정구역'][:-1]
# 서대문구, 서귀포시 같이 이름이 3 자 이상인 경우에 작은 글자로 표시
        if len(dispname.splitlines()[-1]) >= 3:
          fontsize, linespacing = 9.5, 1.5
        else:
          fontsize, linespacing = 11, 1.2
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
               fontsize=fontsize, ha='center', va='center', color=annocolor,
               linespacing=linespacing)
        
#시도경계그린다
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=4)

    plt.gca().invert_yaxis()
    #plt.gca().set_aspect(1)
    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    
    plt.show()
````

### timeseriessplit+튜닝

- 예측을 수행하기에 앞서 하이퍼파라미터 최적화에 이용되는 gridsearchcv함수에 대해 알아보자. gridsearchcv함수 안에는 cv라는 파라미터가 있고, 이것을 숫자로 지정하면, 데이터를 랜덤하게 train,test데이터로 지정하는 교차검증을 수행한다.
하지만, 데이터에 시계열 정보가 있게 되면, 시간 순으로 나열된 데이터의 순서를 보존하여 나뉘어진 train,test데이터로 교차검증이 이루어져야 한다. 
- 참고한 논문에 따르면 시계열 데이터에는 cross validation보다 forward validation을 적용해야함을 알 수 있었고 이 기술을 사이킷런의 timeseriessplit함수를 통해 구현가능함을 알 수 있었다. 

````python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import TimeSeriesSplit
n_splits=4
tscv=TimeSeriesSplit(n_splits)
````

사이킷런에서 TimeSeriesSplit함수를 tscv란 이름으로 불러왔습니다. 
예시로 timeseriessplit함수를 n_splits을 4로 지정하고 12개의 데이터에 적용한 결과는 다음과 같다.

````python
for train, test in tscv.split(range(12)):
    print("train:", train, "test:", test)
````

시간적인 순서를 지키며 데이터가 train,test 데이터로 나뉘어짐을 확인 할 수 있다. 이를 바탕으로 gridsearchcv의 cv파라미터에  tscv를 지정하여 cross validation 대신 forward validation을 이용하여 교차검증을 수행하였다. 

- lasso least angle regression은 L1규제를 선형회귀에 적용한 회귀모델이다.
- 변수선택방법에서 기존 lasso와 차이를 보이는데, 각 단계에서 목표변수와 가장 높은
상관관계가 있는 변수들을 찾고, 변수의 계수를 최소제곱값으로 감소시키는 방식으로 계수를 찾는다.
alpha값이 클수록 과적합을 개선할 수 있으며, 작을 수록 잔차를 개선할 수 있다는 특징이 있다.

````python
from sklearn.linear_model import LassoLars
alphas = np.arange(0,3,0.05)
llar=LassoLars(normalize=False)
parameters_llar={
    'alpha':alphas
}
grid_llar=GridSearchCV(llar,param_grid=parameters_llar,cv=tscv,refit=True,scoring='neg_root_mean_squared_error')
````

````python
grid_llar.fit(X_train,y_train)
print(grid_llar.best_params_)
print(grid_llar.best_score_)
````

````python
y_pred_llar=grid_llar.predict(X_train)
mse_llar=mean_squared_error(y_pred_llar,y_train)
rmse_llar=np.sqrt(mse_llar)
rmse_llar
````

- bayesian ridge는 bayesian regression에서 발전된 모델이다. bayesian regression은
예측변수의 분포를 가우시안으로 가정하여 회귀 계수의 분포를 구합니다. 이때, 회귀계수의 사전분포를 beta분포로 설정한 모델이 
bayesian ridge다.
- 이 모델은 회귀계수를 적합시키는 동안 정규화 파라미터(alpha, lambda)도
각각의 로그 우도를 최대화 하도록 동시에 추정하기 때문에 미리 파라미터 값을 설정하지 않고 데이터에 맞춰 조정되도록 한다는 특징이 있다. 

````python
#bayesian ridge
from sklearn.linear_model import BayesianRidge
br=BayesianRidge()
alpha_init=np.logspace(-5, 0, 50)
lambda_init=np.logspace(-5, 0, 50)

parameters_br={
    'alpha_init':alpha_init,
    'lambda_init':lambda_init
    }
grid_br=GridSearchCV(br,param_grid=parameters_br,cv=tscv,refit=True,scoring='neg_root_mean_squared_error')
````

````python
grid_br.fit(X_train,y_train)
print(grid_br.best_params_)
print(grid_br.best_score_)
````

````python
y_pred_br=grid_br.predict(X_train)
mse_br=mean_squared_error(y_pred_br,y_train)
rmse_br=np.sqrt(mse_br)
rmse_br
````

````python
y_pred_test_br=grid_br.predict(X_test)
mse_test_br=mean_squared_error(y_pred_test_br,y_test)
rmse_test_br=np.sqrt(mse_test_br)
rmse_test_br
````

- Elastic Net 회귀모형은 가중치의 절대값의 합과 제곱합을 동시에 제약 조건으로 가지는 모형으로, Ridge의 L2, Lasso의 L1을 합쳐서 비용함수를 계산한 것이다 alpha가 0에 가까우면 elasticNet은 Ridge에 가까워지고, alpha가 1에 가까우면 elasticNet은 Lasso에 가까워지게 된다.
- elasticNet을 사용하게 되면 Ridge처럼 coefficient의 크기를 줄이면서 Lasso처럼 특정 coefficient를 0으로 만들어 버리면서 특정 독립변수를 사라지게 하는 feature selection 효과를 누릴 수 있다.

````python
# ElasticNet
from sklearn.linear_model import ElasticNet

elasticnet = ElasticNet()
alphas = np.logspace(-4, 0, 200) 
parameters_en = {'alpha': alphas }

grid_en = GridSearchCV(elasticnet, param_grid=parameters_en, cv=tscv, refit=True, scoring='neg_mean_squared_error')
grid_en.fit(X_train, y_train)
print(grid_en.best_params_)
print(grid_en.best_score_)
````

````python
y_pred_en = grid_en.predict(X_train)
mse_en = mean_squared_error(y_pred_en, y_train)
rmse_en = np.sqrt(mse_en)
rmse_en
````

최적 파라미터 값으로 train 데이터를 fit시킨 결과, rmse의 값은 다음과 같다.

````python
y_pred_test_en = grid_en.predict(X_test)
mse_test_en = mean_squared_error(y_pred_test_en, y_test)
rmse_test_en = np.sqrt(mse_test_en)
rmse_test_en
````

최적 파라미터 값으로 test 데이터를 fit한 결과, rmse의 값은 다음과 같다.

````python
# 가중평균 앙상블
y_pred_ensemble = (y_pred_test_br+y_pred_test_llar+y_pred_test_en)/3
mse_ensemble = mean_squared_error(y_pred_ensemble, y_test)
rmse_ensemble = np.sqrt(mse_ensemble)
rmse_ensemble
````

top 3 모델의 predict 값을 산술평균하여 새로운 predict 값을 만들었다. rmse를 계산한 결과 elastic net의 성능에 약간 못미치는 결과를 얻었다.


````python
y_pred_ensemble_w = y_pred_test_br*0.25+y_pred_test_llar*0.25+y_pred_test_en*0.5
mse_ensemble_w = mean_squared_error(y_pred_ensemble_w, y_test)
rmse_ensemble_w = np.sqrt(mse_ensemble_w)
rmse_ensemble_w
````

top1 모델의 가중치를 50%, 나머지 모델의 가중치를 25%로 가중평균하여 새로운 predict 값을 만들었고, rmse를 계산한 결과 역시 elastic net의 성능에 약간 못미치는 결과를 얻었다.
